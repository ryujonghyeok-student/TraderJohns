```python
import pandas as pd

# FEDFUNDS: Federal Funds Effective Rate
# DFEDTARU: Federal Funds Target Range - Upper Limit
# CPIAUCSL: Consumer Price Index for All Urban Consumers: All Items in U.S. City Average
# DGS10: Market Yield on U.S. Treasury Securities at 10-Year Constant Maturity, Quoted on an Investment Basis


# access CPI.csv file in 'data' folder
fed = pd.read_csv('data/FED_rates.csv')
cpi = pd.read_csv('data/CPI.csv')
treasury = pd.read_csv('data/10year_treasury.csv')

fed.index = pd.to_datetime(fed['DATE'])
cpi.index = pd.to_datetime(cpi['DATE'])
treasury.index = pd.to_datetime(treasury['DATE'])

# set start and end date with dataframe data type
start = pd.to_datetime('2023-11-01')
end = pd.to_datetime('2024-01-21')

# combine fed, cpi, treasury using merge from start to end
df = fed.merge(cpi, how='outer', left_index=True, right_index=True)
df = df.merge(treasury, how='outer', left_index=True, right_index=True)
df = df.drop(columns=['DATE_x', 'DATE_y', 'DATE'])
df = df.loc[start:end]

df
```
